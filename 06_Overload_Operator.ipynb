# **Overload Operator in Python**

Operator overloading is used to customize the function of an operator (e.g., **+,*,<,==** etc.) for a user-defined class. It is necessary to overload the operator we want to use with the user-defined data type, without it, the compiler does not know which variables of the user-defined type to add, multiply, or compare.

In Python, overloading is achieved by overriding the method which is specifically for that operator, in the user-defined class. For example, **__add__(self, x)** is a method reserved for overloading **+** operator, and **__eq__(self, x)** is for overloading **==**.

## Chaining

In the context of operator overloading, chaining is when you use an operator multiple times in the same line:



```
# Passing salary as argument to constructor:
a = Person(5000)
b = Person(10000)
c = Person(15000)

# Using '+' two times:
a + b + c
```
To make an overloaded operator capable of performing chaining, we usually return the same user-defined type in the overloading function:



```
def __add__(self, x):
    return Person(self.salary + x.salary)
```

### Example: Create a Person class with attributs name and salary. Overload the operator + and call it in the main with p3 = p1 + p2.

````python
class Person:
    def __init__(self, user_name, user_salary):
        self.name = user_name
        self.salary = user_salary

    def set_name(self, user_name):
        self.name = user_name
    def get_name(self):
        return self.name
    def set_salary(self, user_salary):
        self.salary = user_salary
    def get_salary(self):
        return self.salary

    def __add__(self, other):
        p3 = Person(self.name + ", " + other.name, self.get_salary() + other.get_salary())
        return p3
    def toString(self):
        print("name: ",self.name, ",", "salary:",self.salary)
````

````python
if __name__=="__main__":
    p1 = Person("Faye", 1000)
    p2 = Person("Kevin", 2000)
    p3 = p1 + p2
    p1.toString()
    p2.toString()
    p3.toString()
````

#### Exercice 1: Create an object Fruit that have name, weight and the operators == and + according to the weight. Call it in the main and observe the behavior.

````python
class Fruit:
    def __init__(self, fruit_name, fruit_weight):
        self.name = fruit_name
        self.weight = fruit_weight

    def set_name(self, fruit_name):
        self.name = fruit_name
    def get_name(self):
        return self.name
    def set_weight(self, fruit_weight):
        self.weight = fruit_weight
    def get_weight(self):
        return self.weight

    def __eq__(self, other):
        if (self.get_weight() == other.get_weight()):
            if (self.get_name() == other.get_name()):
                return True #name and weight are equal
            else :
                return False
        else: 
            return False
    def __add__(self, other):
        f3 = Fruit(self.get_name() +", "+ other.get_name(), self.get_weight() + other.get_weight())
        return f3
    def toString(self):
        print("fruit name:", self.get_name(),", weight:", self.get_weight())
````

````python
if __name__=="__main__":
    f1 = Fruit("apple", 50)
    f2 = Fruit("apple", 50)
    f3 = f1 + f2
    f1.toString()
    f2.toString()
    f3.toString()
    print(f1==f2)
````

#### Exercice 2: Create an object Complex Number that have a real part and imaginary part. Overload the operators +, x, **, == and call all operators in the main.

````python
#c1 = Complex(i=1,j=2)
class ComplexNumber:
    def __init__(self, real_num, imaginary_num):
        self.real = real_num
        self.imaginary = imaginary_num
    
    #getters/setters
    def set_real(self, real_num):
        self.real = real_num
    def get_real(self):
        return self.real
    def set_imaginary(self, imaginary_num):
        self.imaginary = imaginary_num
    def get_imaginary(self):
        return self.imaginary
    
    #methods
    def toString(self):
        print(self.get_real(), "+", self.get_imaginary(),"i")
        temp = str(self.get_real()) + "+" + str(self.get_imaginary()) + "i"
        print(temp)
    def __add__(self, other):
        new = ComplexNumber(self.get_real()+other.get_real(),other.get_imaginary()+self.get_imaginary())
        return new
    def __mul__(self,other):
        new_num = ComplexNumber(self.get_real()*other.get_real(), self.get_imaginary()*other.get_imaginary())
        return new_num
    def __eq__(self, other):
        if (self.get_real() == other.get_real()):
            if (self.get_imaginary() == other.get_imaginary()):
                return True
            else:
                return False
        else:
            return False
    def __pow__(self, other):
        new_pow = ComplexNumber(self.get_real()**other.get_real(), self.get_imaginary()**other.get_imaginary())
        return new_pow
    

    
````

````python
if __name__=="__main__":
    a = ComplexNumber(10,4)
    b = ComplexNumber(2,3)

    a.toString()
    b.toString()
    c = a + b
    c.toString()
    d = a * b
    d.toString()
    print(a==b)
    e = a ** b
    e.toString()
````
